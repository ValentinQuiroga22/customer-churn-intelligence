```python
# AI Customer Churn Intelligence
#
# Objective:
# Generate AI-powered explanations and
# retention recommendations for customers
# identified as high churn risk.
```

```python
import pandas as pd

df = pd.read_csv(
    "../data/processed/customer_risk_scoring_business.csv"
)

print(df.columns.tolist()[:10])
```

```python
high_risk = df[
    df["RiskLevel"] == "High"
]

high_risk.shape
```

```python
high_risk.head(5)
```

```python
high_risk.iloc[0]
```

```python
# AI CHURN EXPLANATION ENGINE V1

def explain_customer(customer):
    
    reasons = []
    
    # Contract
    if customer["Contract"] == "Month-to-month":
        reasons.append(
            "Month-to-month contract"
        )
        
    # Tenure
    if customer["tenure"] < 12:
        reasons.append(
            "Short customer tenure"
        )
    
    # Internet
    if customer["InternetService"] == "Fiber optic":
        reasons.append(
            "Fiber optic internet service"
        )
     
     # Support
    if customer["TechSupport"] == "No":
        reasons.append(
            "No technical support service"
        )
    
    # Security
    if customer["OnlineSecurity"] == "No":
        reasons.append(
            "No online security service"
        )
    
    return reasons
```

```python
# RETENTION RECOMMENDATION ENGINE

def generate_recommendations(customer):
    
    recommendations = []
    
    if customer["Contract"] == "Month-to-month":
        recommendations.append(
            "Offer discount for 1-year or 2-year contract"
        )
        
    if customer["OnlineSecurity"] == "No":
        recommendations.append(
            "Promote Online Security package"
        )

    if customer["TechSupport"] == "No":
        recommendations.append(
            "Offer Tech Support bundle"
        )

    if customer["tenure"] < 12:
        recommendations.append(
            "Launch onboarding retention campaign"
        )

    if customer["InternetService"] == "Fiber optic":
        recommendations.append(
            "Review customer satisfaction with Fiber Optic service"
        )

    return recommendations
    
```

```python
customer = high_risk.iloc[0]

explain_customer(customer)
```

```python
# BUSINESS REPORT GENERATOR

def generate_report(customer):
    
    reasons = explain_customer(customer)
    
    recommendations = generate_recommendations(customer)
    
    report = f"""
CUSTOMER CHURN RISK REPORT

Customer ID: {customer["customerID"]}

Risk Score: {customer["RiskScore"]:.1%}
Risk Level: {customer["RiskLevel"]}

Main Risk Factors:
"""
    for reason in reasons:
        report += f"\n- {reason}"
    
    report += "\n\nRECOMMENDED ACTIONS:\n"
    
    for rec in recommendations:
        report += f"\n- {rec}"
    
    return report 
```

```python
generate_report(high_risk.iloc[0])
```

```python
for i in range (3):
    print(generate_report(high_risk.iloc[i]))
    print("\n" + "="*80 + "\n")
```

```python
# EXECUTIVE SUMMARY GENERATOR

def generate_executive_summary(customer):
    
    reasons = explain_customer(customer)
    
    summary = (
        f"Customer {customer["customerID"]} "
        f"has been classified as {customer["RiskLevel"]} Risk "
        f"with a predicted churn probability of "
        f"{customer["RiskScore"]:.1%}. "
    )
    
    if len(reasons) > 0:
        
        summary += (
            "The primary risk factors include "
            + ", ".join(reasons[:-1])
        )
        
        if len(reasons) > 1:
            summary += f" and {reasons[-1]}."
        
        else:
            summary += f"{reasons[0]}"
        
    summary += (
        " Based on historical customer behavior, "
        "similar profiles have shown an elevated "
        "probability of churn. "
        "A targeted retention strategy is recommended."
    )

    return summary
```

```python
generate_executive_summary(
    high_risk.iloc[0]
)
```

```python
# Top 10 high risk customers

top_10 = high_risk.sort_values(
    by="RiskScore",
    ascending=False
). head(10)

top_10[[
    "customerID",
    "RiskScore",
    "RiskLevel"
]]
```

```python
for i in range(3):
    
    customer = top_10.iloc[i]
    
    print("=" * 80)
    
    print(
        generate_executive_summary(customer)
    )
    print("\n")
```
